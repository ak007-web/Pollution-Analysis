# Label Encoding on Air Quality Column


```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
col=['Air Quality']
for i in col:
    df[i]=le.fit_transform(df[i])
df.head()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=['Air Quality'])
y=df['Air Quality']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Decison Tree Classifier and Confusion Matrix

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_preddt=dt.predict(x_test)
print("Random Forest Classifier\n",classification_report(y_test,y_preddt))
print("Accuracy Score: ",accuracy_score(y_test,y_preddt))

#Confusion Matrix for Decision Tree 
from sklearn import metrics
import matplotlib.pyplot as plt

confusion_matrix = metrics.confusion_matrix(y_test, y_preddt)

unique_labels = sorted(set(y_test) | set(y_preddt))
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=unique_labels)

cm_display.plot()
plt.show()
```

# Random Forest Classification and it's Confusion Matrix

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_predrf=rf.predict(x_test)
print("Random Forest Classifier\n",classification_report(y_test,y_predrf))
print("Accuracy Score: ",accuracy_score(y_test,y_predrf))

#Confusion Matrix
from sklearn import metrics
import matplotlib.pyplot as plt

confusion_matrix = metrics.confusion_matrix(y_test, y_predrf)

unique_labels = sorted(set(y_test) | set(y_predrf))
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=unique_labels)

cm_display.plot()
plt.show()
```
